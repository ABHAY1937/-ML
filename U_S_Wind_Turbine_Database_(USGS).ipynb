The U.S. Wind Turbine Database is a comprehensive database of onshore wind turbines in the United States, which contains detailed information on the location, size, and specifications of individual wind turbines, as well as information on the wind farms and projects they are a part of.
 The database includes turbines that are currently in operation, as well as those that have been retired or decommissioned, making it a valuable resource for analyzing the development and evolution of the wind energy industry in the United States.
The USGS Wind Turbine Database can be used for a wide range of applications, such as site selection, energy planning, and environmental impact assessment, making it a valuable tool for advancing the development and deployment of renewable energy in the United States.

CREATE A DATAFRAME WITH CATEGORICAL VALUES USING GET_DUMMIES

```python
df2=pd.get_dummies(df1[["t_state","t_county","p_name","t_img_srce"]],drop_first=True)
```

COMBINING THESE TWO DATASET 

```python
dfe=pd.concat([df1,df2],axis=1)
```

```python
dfe.drop(columns=["t_state","t_county","p_name","t_img_srce","case_id"],axis=1,inplace=True)
```

CONVERTING OBJECT TO FLOAT

```python
dfe["p_tnum"]=dfe["p_tnum"].astype(float)
dfe["retrofit"]=dfe["retrofit"].astype(float)
dfe["t_conf_atr"]=dfe["t_conf_atr"].astype(float)
dfe["t_conf_loc"]=dfe["t_conf_loc"].astype(float)
dfe["xlong"]=dfe["xlong"].astype(float)
dfe["ylat"]=dfe["ylat"].astype(float)
dfe["t_img_srce_Digital Globe"]=dfe["t_img_srce_Digital Globe"].astype(float)
dfe["t_img_srce_NAIP"]=dfe["t_img_srce_NAIP"].astype(float)
dfe["t_img_srce_Digital Globe"]=dfe["t_img_srce_Digital Globe"].astype(float)
dfe["p_year"]=dfe["p_year"].astype(float)
dfe["t_fips"]=dfe["t_fips"].astype(float)
                                          
```

CONVERTING INT OR FLOAT TO NaN

```python
dfe.loc[dfe.p_tnum==0, "p_tnum"]=np.NaN
dfe.loc[dfe.retrofit==0, "retrofit"]=np.NaN
dfe.loc[dfe.t_conf_atr==0, "t_conf_atr"]=np.NaN
dfe.loc[dfe.t_conf_loc==0, "t_conf_loc"]=np.NaN
dfe.loc[dfe.xlong==0, "xlong"]=np.NaN
dfe.loc[dfe.ylat==0, "ylat"]=np.NaN
dfe.loc[dfe.p_year==0, "p_year"]=np.NaN
```

HANDLING MISSING VALUES

```python
dfe["p_tnum"]=dfe["p_tnum"].fillna(dfe["p_tnum"].mean())
dfe["retrofit"]=dfe["retrofit"].fillna(dfe["retrofit"].mean())
dfe["t_conf_atr"]=dfe["t_conf_atr"].fillna(dfe["t_conf_atr"].mean())
dfe["t_conf_atr"]=dfe["t_conf_atr"].fillna(dfe["t_conf_atr"].mean())
dfe["xlong"]=dfe["xlong"].fillna(dfe["xlong"].mean())
dfe["ylat"]=dfe["ylat"].fillna(dfe["ylat"].mean())
```

K-MEANS CLUSTER

```python
from sklearn.cluster import KMeans
import time

start_time = time.time()
kmeans = KMeans(n_clusters=2, n_init=10)
wcss = []
for i in range(1, 21): # reduce range of clusters to test
  kmeans = KMeans(n_clusters=i, init='k-means++', n_init=5, max_iter=100, random_state=42) # reduce n_init and max_iter
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

```python
#TIME TAKEN FOR EXCECUTING THE CODE
print("--- Execution time: %s seconds ---" % (time.time() - start_time))
```

```python
print(wcss)
```

ELBOW-POINT

```python
import matplotlib.pyplot as plt

plt.plot(range(1,21),wcss)
plt.xlabel("cluster")
plt.ylabel("varients")
plt.title("elbow point graph")
plt.show()
```

```python
Kmeans=KMeans(n_clusters=4,init='k-means++',random_state=42)
y_pred=Kmeans.fit_predict(x)
y_pred
```

VARIABLE SEPERATION USING i.loc

```python
a=x.iloc[:,:-1].values
b=x.iloc[:,-1].values
```

DATA TRAINING AND TESTING

```python
from sklearn.model_selection import train_test_split
a_train ,a_test,b_train,b_test =train_test_split(a,b,test_size=0.30,random_state=42)
a_train.shape
```

```python
a_test.shape
```

DATA PREPROCESSING

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(a_train)
a_train=scalar.fit_transform(a_train)
a_test=scalar.fit_transform(a_test)
```

IMPLIMENTING SUPPORT VECTOR MACHINE ALGORITHM

```python
#model selection
from sklearn.svm import SVC
model=SVC()
model.fit(a_train,b_train)
```

```python
ypred=model.predict(a_test)
ypred
```

ACCURACY SCORE

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
score=accuracy_score(b_test,ypred)
score
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,ConfusionMatrixDisplay
result=confusion_matrix(b_test,ypred)
labels=[0,1,2,3]
cmd=ConfusionMatrixDisplay(result,display_labels=labels)
cmd.plot()
```
